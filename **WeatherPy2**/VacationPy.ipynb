```python
#RUN LOOP FOR HOTEL NAMES RESULTS
for index, row in clearweatherfin.iterrows():
    #MAKE LAT AND LON INTO VARIABLES
    lat = row['Latitude']
    lon = row['Longitude']

    #VARIABLE BASE URL AND PARAMETERS
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {
        "location": f"{lat},{lon}",
        "radius": 5000,
        "types" : "hotel",
        "key": g_key
    }
    
    #REQUEST JSON FOR LAT AND LON IN THE GIVEN ROW
    response = requests.get(base_url, params=params).json()
    #STORE RESULTS
    results = response['results']
    #RUN TRY AND EXPECT FOR FILTERING RESULTS
    try:
        print(f"Closest hotel is {results[0]['name']}.")
        #STORE NAME OF HOTEL INTO 'HOTEL NAME' COLUMN IN DF
        clearweatherfin.loc[index, 'Hotel Name'] = results['results'][0]['name']
    except:
        print("Error: Skip Result")
```
